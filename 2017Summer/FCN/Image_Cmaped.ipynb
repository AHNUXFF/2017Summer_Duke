```python
import scipy.misc as misc
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
%matplotlib inline
import os
```

```python
"""draw train results"""
mode = 'train'
patchsize = 4
imageDir = os.path.join('logs_batch%d' % patchsize,'images')
if(not os.path.isdir(imageDir)):
    print('No')
for imageInd in range(0,10):
    gt = misc.imread(os.path.join(imageDir,'gt_{}{}.png'.format(mode,imageInd)))
    inp = misc.imread(os.path.join(imageDir,'inp_{}{}.png'.format(mode,imageInd)))
    pred = misc.imread(os.path.join(imageDir,'pred_{}{}.png'.format(mode,imageInd)))
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3,figsize=(15,15))
    ax1.imshow(inp)
    ax2.imshow(gt,cmap=cm.gist_stern, vmin=0, vmax=151)
    ax3.imshow(pred,cmap=cm.gist_stern, vmin=0, vmax=151)
```
